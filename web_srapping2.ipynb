```python
i1 = 0
data = []
links = []
count=0;

for i in headings:
    if i1 < 10:
        # Append the heading text to the data list
        data.append(i.text.strip())

        # Find the anchor tag with the class 'w_img'
        link = i.find('a', attrs={'class': "w_img"})


        # Check if the anchor tag exists and has an href attribute
        if link and 'href' in link.attrs:
            # links.append('https://timesofindia.indiatimes.com/link['href'])  # Append the href link to the links list
            links.append('https://timesofindia.indiatimes.com/'+link.get('href'))

        i1 += 1
        count=count+1
    else:
        break

# Print or use the 'links' list which now contains href URLs
# print(links)
# print(count)
```
